```python
import requests
import io
import polars as pl

# API endpoint URL
url = "https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/medan?unitGroup=us&include=hours&key=ESFK258ZHM3EKCNP8H4UHHPGT&contentType=csv"

# Make a GET request to the API endpoint
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Get the content of the response
    content = response.content

    # Create a file-like object from the content
    file_obj = io.BytesIO(content)

    # Read the CSV data using Polars
    df = pl.read_csv(file_obj, dtypes={'sealevelpressure': pl.Float64})


    # Print the dataframe
    print(df)
else:
    print("Failed to retrieve data from the API.")
```

```python
import requests
import sys
import csv
import polars as pl

response = requests.request("GET", "https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/medan?unitGroup=us&include=hours&key=ESFK258ZHM3EKCNP8H4UHHPGT&contentType=csv")
if response.status_code != 200:
    print('Unexpected Status code: ', response.status_code)
    sys.exit()

# Parse the results as CSV
CSVText = response.text

# Convert CSV data to list of rows
rows = list(csv.reader(CSVText.splitlines(), delimiter=',', quotechar='"'))

# Get column names from the first row
column_names = rows[0]

# Get data rows excluding the header
data_rows = rows[1:]

# Create a polars DataFrame with specified column names
df = pl.DataFrame(data_rows, columns=column_names)

# Print the DataFrame
print(df)
```

```python
import polars as pl
df= pl.read_csv('https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/jakarta/2023-06-18/2023-06-18?unitGroup=us&include=hours&key=ESFK258ZHM3EKCNP8H4UHHPGT&contentType=csv',infer_schema_length=1000)
df= (
    df
    .with_columns(
        pl.col('datetime')
        .str.strptime(pl.Datetime,
                      fmt="%Y-%m-%dT%H:%M:%S%.f",
                      strict=False)
    )
    .drop('name')
)
df.head()
```
